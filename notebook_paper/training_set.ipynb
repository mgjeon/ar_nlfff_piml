```python
from pathlib import Path 
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_path = sorted(list(Path("/mnt/f/isee_dataset/").glob("*/input/*.npz")))
```

```python
len(input_path)
```

```python
train_input_path = [f for f in input_path if not ("11158" in str(f)) and not ("12673" in str(f))]
```

```python
len(train_input_path)
```

```python
test_input_path = [f for f in input_path if ("11158" in str(f)) or ("12673" in str(f))]
```

```python
len(test_input_path)
```

```python
inputs_max_train = [np.max(np.load(f)['input']) for f in train_input_path]
inputs_min_train = [np.min(np.load(f)['input']) for f in train_input_path]
```

```python
plt.plot(inputs_max_train)
```

```python
np.mean(inputs_max_train), np.median(inputs_max_train)
```

```python
plt.plot(inputs_min_train)
```

```python
np.mean(inputs_min_train), np.median(inputs_min_train)
```

```python
inputs_max_test = [np.max(np.load(f)['input']) for f in test_input_path]
inputs_min_test = [np.min(np.load(f)['input']) for f in test_input_path]
```

```python
plt.plot(inputs_max_test)
```

```python
np.mean(inputs_max_test), np.median(inputs_max_test)
```

```python
plt.plot(inputs_min_test)
```

```python
np.mean(inputs_min_test), np.median(inputs_min_test)
```
