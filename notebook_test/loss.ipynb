```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('train_loss.csv')
val_df = pd.read_csv('val_loss.csv')
```

```python
train_x = train_df['Step']
train_y = train_df['Value']

val_x = val_df['Step']
val_y = val_df['Value']
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(train_x, train_y, label='train loss')
plt.plot(val_x, val_y, label='val loss')
plt.legend()
plt.show()
```
